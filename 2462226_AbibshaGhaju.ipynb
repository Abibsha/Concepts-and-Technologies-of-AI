**Problem 1**

**Basic Data Exploration & Trend Visualization**

**3.1	Problem - 1A - Single Year HDI Exploration (Latest Year: 2022)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Concepts and Technologies of AI/Human_Development_Index_Dataset.csv', encoding='latin-1')
```

```python
# Task 1: Extract Latest Year
unique_years = df['year'].unique()
print("Unique years in the dataset are given below:\n", unique_years)

hdi_2022_df = df[df['year'] == 2022].copy()
print("\nShape of 2022 dataset is given below:\n", hdi_2022_df.shape)
```

```python
# Task 2: Data Exploration
print("First 10 rows of the 2022 dataset is given below:")
display(hdi_2022_df.head(10))

print(f"\nThe number of rows in the 2022 dataset is {hdi_2022_df.shape[0]}.")
print(f"The number of columns in the 2022 dataset is {hdi_2022_df.shape[1]}.")

print("\nAll column names and their data types are given below:\n")
display(hdi_2022_df.dtypes)
```

```python
# Task 3: Missing Values and Data Cleaning
print("Missing values in each column:")
print(hdi_2022_df.isnull().sum())

duplicate_count = hdi_2022_df.duplicated().sum()
print(f"\nNumber of duplicate rows found: {duplicate_count}")

object_columns = hdi_2022_df.select_dtypes(include='object').columns
special_char_count = hdi_2022_df[object_columns].isin(['-', 'NA', 'N/A']).sum().sum()
print(f"Number of values with special characters representing missing data: {special_char_count}")

numeric_columns = ['hdi', 'life_expectancy', 'gross_inc_percap']
for col in numeric_columns:
    hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')

hdi_2022_df.dropna(subset=['hdi'], inplace=True)
```

```python
# Task 4: Basic Statistics
mean_hdi = hdi_2022_df['hdi'].mean()
median_hdi = hdi_2022_df['hdi'].median()
std_hdi = hdi_2022_df['hdi'].std()

print(f"Mean of HDI for the year 2022: {mean_hdi}")
print(f"Median of HDI for the year 2022: {median_hdi}")
print(f"Standard Deviation of HDI for the year 2022: {std_hdi}")

highest_hdi_country = hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax()]
print("\nCountry with the Highest HDI in 2022:")
display(highest_hdi_country[['country', 'hdi']])

lowest_hdi_country = hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin()]
print("\nCountry with the Lowest HDI in 2022:")
display(lowest_hdi_country[['country', 'hdi']])
```

```python
# Task 5: Filtering and Sorting
high_hdi_df = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]

high_hdi_sorted = high_hdi_df.sort_values(by='gross_inc_percap', ascending=False)

print("Top 10 countries with HDI greater than 0.800 sorted by GNI per capita are given below:")
display(high_hdi_sorted.head(10)[['country', 'hdi', 'gross_inc_percap']].head(10))
```

```python
# Task 6: Adding HDI Category Column
def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi < 0.700:
        return "Medium"
    elif 0.700 <= hdi < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(classify_hdi)

print("HDI Category distribution is given below:")
display(hdi_2022_df['HDI Category'].value_counts())
display(hdi_2022_df[['country', 'hdi', 'HDI Category']].head())

hdi_2022_df.to_csv('/content/drive/MyDrive/Concepts and Technologies of AI/HDI_category_added.csv',index=False)
print("New dataset HDI_category_added.csv created.")
```

**3.2	Problem - 1B - HDI Visualization and Trend Analysis (2020 â€“ 2022)**

```python
# Task 1: Data Extraction and Saving:
hdi_1B = df[df['year'].isin([2020, 2021, 2022])].copy()
hdi_1B.to_csv("/content/drive/MyDrive/Concepts and Technologies of AI/HDI_problem_1B.csv", index=False)

print("New dataset HDI_problem_1B.csv created successfully.")
```

```python
# Task 2: Data Cleaning
print("Missing values before cleaning:")
print(hdi_1B[['hdi', 'country', 'year']].isnull().sum())

hdi_1B['country'] = hdi_1B['country'].astype(str).str.strip().str.title()
hdi_1B.replace(['â€“', '-', 'NA', 'N/A', ''], np.nan, inplace=True)

numeric_cols = ['hdi', 'life_expectancy', 'gross_inc_percap']
for col in numeric_cols:
    hdi_1B[col] = pd.to_numeric(hdi_1B[col], errors='coerce')

duplicate_count = hdi_1B.duplicated().sum()
print(f"\nNumber of duplicate rows found: {duplicate_count}")
hdi_1B.drop_duplicates(inplace=True)

hdi_1B.dropna(subset=['hdi', 'country', 'year'], inplace=True)

hdi_1B['year'] = hdi_1B['year'].astype(int)

print("\nMissing values after cleaning:")
print(hdi_1B[['hdi', 'country', 'year']].isnull().sum())
```

```python
# Task 3: Visualization Tasks:
# A. Line Chart â€” HDI Trend (Country-Level):
selected_countries = ['Nepal', 'India', 'Bangladesh', 'Sri Lanka', 'Pakistan']
country_trend_df = hdi_1B[hdi_1B['country'].isin(selected_countries)]

plt.figure(figsize=(6,4))

for country in selected_countries:
    data = country_trend_df[country_trend_df['country'] == country]
    plt.plot(data['year'], data['hdi'], marker='o', label=country)

plt.title("HDI Trend (2020â€“2022) for Selected Countries")
plt.xlabel("Year")
plt.ylabel("HDI")
plt.legend()
plt.grid(True)
plt.xticks([2020, 2021, 2022])
plt.figtext(0.5, -0.10,"Fig 1: HDI trends for five countries from 2020 to 2022.",ha="center", fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# B. Generate Visualizations:
#Bar Chart: Average HDI by Region (2020â€“2022)
region_year_hdi = (country_trend_df.groupby(['country', 'year'], as_index=False)['hdi'].mean())

plt.figure(figsize=(8,5))
sns.barplot(data=region_year_hdi, x='country', y='hdi', hue='year', palette='Pastel1')
plt.title("Average HDI by Region (2020â€“2022)")
plt.xlabel("Region")
plt.ylabel("Average HDI")
plt.ylim(0, 1)
plt.xticks(rotation=30)
plt.legend(title="Year")
plt.figtext(0.5, -0.10,"Fig 2: Comparison of average HDI across regions from 2020 to 2022.",ha="center", fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# Box Plot: HDI Distribution for 2020, 2021, and 2022
hdi_distribution_df = hdi_1B[hdi_1B['year'].isin([2020, 2021, 2022])]
plt.figure(figsize=(6,4))
sns.boxplot(data=hdi_distribution_df, x='year', y='hdi')
plt.title("Distribution of HDI for 2020, 2021, and 2022")
plt.xlabel("Year")
plt.ylabel("HDI")
plt.ylim(0,1)
plt.figtext(0.5, -0.10,"Fig 3: Distribution of HDI across years.",ha="center", fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: HDI vs. GNI per Capita
if 'gross_inc_percap' in hdi_1B.columns:
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=hdi_1B, x='gross_inc_percap', y='hdi')
    plt.title("HDI vs GNI per Capita (2020â€“2022)")
    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.figtext(0.5, -0.10,"Fig 4: HDI vs GNI per Capita (2020â€“2022)",ha="center", fontsize=9)
    plt.show()
else:
    print("GNI per Capita variable not available in the dataset.")
```

**Problem 2**

**Advanced HDI Exploration**


```python
# Task 1: Create South Asia Subset
south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

south_asia_df = df[df['country'].isin(south_asia_countries)].copy()
south_asia_df.to_csv("/content/drive/MyDrive/Concepts and Technologies of AI/HDI_SouthAsia.csv", index=False)

print("New dataset HDI_SouthAsia.csv created successfully.")
```

```python
# Task 2: Composite Development Score
south_asia_df["Composite Score"] = (0.30 * south_asia_df["life_expectancy"] + 0.30 * south_asia_df["gross_inc_percap"])

country_composite = (south_asia_df.groupby("country", as_index=False).agg({"Composite Score": "mean", "hdi": "mean"}))
composite_rank = country_composite.sort_values(by="Composite Score", ascending=False)
top5 = composite_rank.head(5)

plt.figure(figsize=(8,5))
plt.barh(top5["country"], top5["Composite Score"])
plt.title("Top 5 South Asian Countries by Composite Score")
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.figtext(0.5, -0.10,"Fig 5: Composite Development Score of top 5 countries",ha="center", fontsize=9)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Task 3: Outlier Detection
def outlier_detection(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

south_asia_df["HDI_outlier"] = outlier_detection(south_asia_df["hdi"])
south_asia_df["GNI_outlier"] = outlier_detection(south_asia_df["gross_inc_percap"])
south_asia_df["Outlier"] = south_asia_df["HDI_outlier"] | south_asia_df["GNI_outlier"]

plt.figure(figsize=(6,4))
sns.scatterplot(data=south_asia_df,x="gross_inc_percap",y="hdi",hue="Outlier",palette={False: "red", True: "green"})
plt.title("GNI per Capita vs HDI")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.figtext(0.5, -0.10,"Fig 6: Outlier Detection in HDI and GNI per Capita ",ha="center", fontsize=9)
plt.ylim(0,1)
plt.tight_layout()
plt.show()
```

```python
# Task 4: Exploring Metric Relationships
metrics = ["gender_development", "life_expectancy"]

for metric in metrics:
    corr = south_asia_df[metric].corr(south_asia_df["hdi"])
    print(f"Pearson correlation between {metric.replace('_', ' ').title()} and HDI: {corr:.3f}")

for metric in metrics:
    plt.figure(figsize=(6,5))
    sns.regplot(data=south_asia_df, x=metric, y="hdi", scatter_kws={"alpha":0.6, "color":"blue"}, line_kws={"color":"red"})
    plt.title(f"HDI vs {metric.replace('_', ' ').title()}")
    plt.xlabel(metric.replace("_", " ").title())
    plt.ylabel("HDI")
    plt.ylim(0,1)
    plt.figtext(0.5, -0.10, f"Fig 7: Relationship between {metric.replace('_', ' ').title()} and HDI", ha="center", fontsize=9)
    plt.tight_layout()
    plt.show()
```

```python
# Task 5: Gap Analysis
south_asia_df["GNI_HDI_Gap"] = south_asia_df["gross_inc_percap"] - south_asia_df["hdi"]

gap_rank_desc = south_asia_df.groupby("country", as_index=False)["GNI_HDI_Gap"].mean().sort_values(by="GNI_HDI_Gap", ascending=False)
gap_rank_asc = gap_rank_desc.sort_values(by="GNI_HDI_Gap", ascending=True)

top3_positive = gap_rank_desc.head(3)
plt.figure(figsize=(6,4))
plt.barh(top3_positive["country"], top3_positive["GNI_HDI_Gap"], color='orange')
plt.title("Top 3 Positive GNI-HDI Gaps")
plt.xlabel("GNI-HDI Gap")
plt.ylabel("Country")
plt.figtext(0.5, -0.10, "Fig 8: Countries where GNI is higher than expected relative to HDI", ha="center", fontsize=9)
plt.show()

top3_negative = gap_rank_asc.head(3)
plt.figure(figsize=(6,4))
plt.barh(top3_negative["country"], top3_negative["GNI_HDI_Gap"], color='purple')
plt.title("Top 3 Negative GNI-HDI Gaps")
plt.xlabel("GNI-HDI Gap")
plt.ylabel("Country")
plt.figtext(0.5, -0.10, "Fig 9: Countries where HDI is higher than expected relative to GNI", ha="center", fontsize=9)
plt.show()
```

**Problem 3**

**Comparative Regional Analysis: South Asia vs Middle East**

```python
# Task 1: Create Middle East Subset:
south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
middle_east_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
                         "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]

sa_2020_22 = hdi_1B[hdi_1B['country'].isin(south_asia_countries)].copy()
me_2020_22 = hdi_1B[hdi_1B['country'].isin(middle_east_countries)].copy()

sa_2020_22.to_csv("/content/drive/MyDrive/Concepts and Technologies of AI/HDI_SouthAsia_2020_2022.csv", index=False)
me_2020_22.to_csv("/content/drive/MyDrive/Concepts and Technologies of AI/HDI_MiddleEast_2020_2022.csv", index=False)

print("New datasets saved successfully.")
```

```python
# Task 2: Descriptive Statistics
sa_hdi_mean = sa_2020_22['hdi'].mean()
sa_hdi_std = sa_2020_22['hdi'].std()
me_hdi_mean = me_2020_22['hdi'].mean()
me_hdi_std = me_2020_22['hdi'].std()

print(f"South Asia HDI: Mean={sa_hdi_mean:.3f}, Std={sa_hdi_std:.3f}")
print(f"Middle East HDI: Mean={me_hdi_mean:.3f}, Std={me_hdi_std:.3f}")

better_region = "South Asia" if sa_hdi_mean > me_hdi_mean else "Middle East"
print(f"The region with higher average HDI is: {better_region}")
```

```python
# Task 3: Top and Bottom Performers
def top_bottom_performers(region_df, region_label):
    avg_hdi = region_df.groupby("country")["hdi"].mean().reset_index()

    highest = avg_hdi.sort_values("hdi", ascending=False).iloc[:3]
    lowest = avg_hdi.sort_values("hdi").iloc[:3]

    display(f"{region_label} : Top 3 Countries", highest)
    display(f"{region_label} : Bottom 3 Countries", lowest)

    return highest, lowest

sa_high, sa_low = top_bottom_performers(sa_2020_22, "South Asia")
me_high, me_low = top_bottom_performers(me_2020_22, "Middle East")

plot_df = pd.concat([
    sa_high.assign(area="South Asia"),
    me_high.assign(area="Middle East"),
    sa_low.assign(area="South Asia"),
    me_low.assign(area="Middle East")
])

plt.figure(figsize=(10, 6))
sns.barplot(data=plot_df, x="country", y="hdi", hue="area", palette="Pastel1")
plt.title("HDI Comparison: Top and Bottom Countries by Region")
plt.ylabel("Average HDI")
plt.figtext(0.5, -0.10, "Fig 10: Comparing top and bottom performers across two region", ha="center", fontsize=9)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Task 4: Metric Comparisons:
metric_list = ["gender_development", "life_expectancy", "gross_inc_percap"]

sa_avg_metrics = (sa_2020_22[metric_list].mean(numeric_only=True).to_frame().T.assign(Area="South Asia"))
me_avg_metrics = (me_2020_22[metric_list].mean(numeric_only=True).to_frame().T.assign(Area="Middle East"))

merged_long = pd.concat([sa_avg_metrics, me_avg_metrics]).melt(id_vars="Area",var_name="Metric_Name",value_name="Avg_Value")

plt.figure(figsize=(8, 6))
sns.barplot(data=merged_long, x="Metric_Name", y="Avg_Value", hue="Area", palette="Pastel2")
plt.title("Comparison of Metrics Across Regions")
plt.ylabel("Average Value")
plt.figtext(0.5, -0.10, "Fig 11: Comparison of metrics across regions", ha="center", fontsize=9)
#plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Task 5: HDI Disparity
sa_hdi_max = sa_2020_22["hdi"].max()
sa_hdi_min = sa_2020_22["hdi"].min()
sa_hdi_range = sa_hdi_max - sa_hdi_min
sa_hdi_cv = sa_2020_22["hdi"].std() / sa_2020_22["hdi"].mean()

me_hdi_max = me_2020_22["hdi"].max()
me_hdi_min = me_2020_22["hdi"].min()
me_hdi_range = me_hdi_max - me_hdi_min
me_hdi_cv = me_2020_22["hdi"].std() / me_2020_22["hdi"].mean()

print("South Asia HDI Range:", sa_hdi_range)
print("South Asia HDI Coefficient of Variation:", sa_hdi_cv)

print("\nMiddle East HDI Range:", me_hdi_range)
print("Middle East HDI Coefficient of Variation:", me_hdi_cv)

if sa_hdi_cv > me_hdi_cv:
    print("\nSouth Asia exhibits more variation in HDI.")
else:
    print("\nMiddle East exhibits more variation in HDI.")
```

```python
# Task 6: Correlation Analysis
sa_corr_gender = sa_2020_22["hdi"].corr(sa_2020_22["gender_development"])
sa_corr_life = sa_2020_22["hdi"].corr(sa_2020_22["life_expectancy"])

me_corr_gender = me_2020_22["hdi"].corr(me_2020_22["gender_development"])
me_corr_life = me_2020_22["hdi"].corr(me_2020_22["life_expectancy"])

print("South Asia Correlations:")
print("HDI vs Gender Development Index:", sa_corr_gender)
print("HDI vs Life Expectancy Index:", sa_corr_life)

print("\nMiddle East Correlations:")
print("HDI vs Gender Development Index:", me_corr_gender)
print("HDI vs Life Expectancy Index:", me_corr_life)

plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
sns.regplot(data=sa_2020_22,x="gender_development",y="hdi")
plt.title("South Asia: HDI vs Gender Development Index")

plt.subplot(2, 2, 2)
sns.regplot(data=sa_2020_22,x="life_expectancy",y="hdi")
plt.title("South Asia: HDI vs Life Expectancy Index")

plt.subplot(2, 2, 3)
sns.regplot(data=me_2020_22,x="gender_development",y="hdi")
plt.title("Middle East: HDI vs Gender Development Index")

plt.subplot(2, 2, 4)
sns.regplot(data=me_2020_22,x="life_expectancy",y="hdi")
plt.title("Middle East: HDI vs Life Expectancy Index")
plt.figtext(0.5, -0.01, "Fig 12: Scatter plots with trendlines showing the correlation of HDI with Gender Development Index and Life Expectancy Index for South Asia and Middle East.",
            wrap=True, ha='center', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Task 7: Outlier Detection
def find_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    return (series < lower_limit) | (series > upper_limit)

for region_label, df in [("South Asia", sa_2020_22), ("Middle East", me_2020_22)]:
    df_copy = df.copy()
    df_copy["HDI_Outlier"] = find_outliers(df_copy["hdi"])
    df_copy["GNI_Outlier"] = find_outliers(df_copy["gross_inc_percap"])
    df_copy["Any_Outlier"] = df_copy["HDI_Outlier"] | df_copy["GNI_Outlier"]

    print(f"\n{region_label} Outliers:")
    display(df_copy[df_copy["Any_Outlier"]][["country", "year", "hdi", "gross_inc_percap"]])

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(sa_2020_22.index, sa_2020_22["hdi"], color="blue", label="Normal")
plt.scatter(
    sa_2020_22[find_outliers(sa_2020_22["hdi"])].index,
    sa_2020_22[find_outliers(sa_2020_22["hdi"])]["hdi"],
    color="red", label="Outlier"
)
plt.title("South Asia: HDI Outliers")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(me_2020_22.index, me_2020_22["hdi"], color="green", label="Normal")
plt.scatter(
    me_2020_22[find_outliers(me_2020_22["hdi"])].index,
    me_2020_22[find_outliers(me_2020_22["hdi"])]["hdi"],
    color="red", label="Outlier"
)
plt.title("Middle East: HDI Outliers")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.legend()
plt.tight_layout()
plt.figtext(0.5, -0.02,"Fig 13: Scatter plots showing HDI outliers in South Asia and Middle East (red points)",
            wrap=True, horizontalalignment='center', fontsize=10
)
plt.show()
```
