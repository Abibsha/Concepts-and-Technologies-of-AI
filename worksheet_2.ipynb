**Problem 2 - Data Imputations:**

Complete all the following Task:

• Dataset for the Task: "medical_student.csv"


1.	Load the provided dataset and import in pandas DataFrame.

```python
med_dataset = pd.read_csv("/content/drive/MyDrive/Concepts and Technologies of AI/medical_students_dataset.csv")
```

2.	Check info of the DataFrame and identify column with missing (null) values.

```python
med_dataset.info()
```

```python
print("Columns with missing null values are given below:\n", med_dataset.isnull().sum())
```

3.	For the column with missing values fill the values using various techniques we discussed above.

```python
for col in med_dataset.columns:
    if med_dataset[col].dtype != 'object':
        med_dataset[col] = med_dataset[col].fillna(med_dataset[col].median())
        print(f"The missing values in {col} is filled using median technique.")
    else:
        med_dataset[col] = med_dataset[col].fillna(med_dataset[col].mode()[0])
        print(f"The missing values in {col} is filled using mode technique.")
```

4.	Check for any duplicate values present in Dataset and do necessary to manage the duplicate items. {Hint: dataset.duplicated.sum()}

```python
print("Duplicate values present in dataset is:\n", med_dataset.duplicated().sum())
```

```python
med_dataset = med_dataset.drop_duplicates()
print("Duplicate values are removed from dataset.")
```

**Problem - 4:**

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visualization to show how the survival distributions vary by gender.


```python
import matplotlib.pyplot as plt
```

```python
gender_survival = titanic_dataset.groupby("Sex")['Survived'].mean()
```

```python
print(gender_survival)
```

```python
plt.bar(gender_survival.index, gender_survival.values, color=["pink", "skyblue"])
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Rate of Survival")
plt.show()
```

**Problem - 5:**

Draw a visualization that breaks your visualization from Exercise 3 down by the port of embarkation (’Embarked’). In this instance, compare the ports ’C’ (Cherbourg), ’Q’ (Queenstown), and ’S’ (Southampton).
