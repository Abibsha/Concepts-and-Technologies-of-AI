**Problem - 4: Matrix Operations:**

 • For the following arrays:

A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);

Prove following with Numpy:

1.	Prove A.A−1 = I.
2.	Prove AB ̸= BA.
3.	Prove (AB)T = BTAT.
•	Solve the following system of Linear equation using Inverse Methods.
2x − 3y + z = −1 x − y + 2z = −3 3x + y − z = 9
{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
•	Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function of Numpy}


```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. A * A^-1 = I
A_inverse = np.linalg.inv(A)
print("A * A^-1:\n", np.dot(A, A_inverse))
```

```python
# 2. AB != BA
print("AB:\n", np.dot(A, B))
print("BA:\n", np.dot(B, A))
```

```python
# 3. (AB)^T = B^T * A^T
AB = np.dot(A, B)
lhs = AB.T
rhs = np.dot(B.T, A.T)
print("(AB)^T:\n", lhs)
print("B^T A^T:\n", rhs)
```

```python
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3,  1, -1]])

B = np.array([-1, -3, 9])
A_inverse = np.linalg.inv(A)
X = np.dot(A_inverse, B)
print("Solution using inverse:\n", X)
```

**4.2	Experiment: How Fast is Numpy?**
1. Element-wise Addition:

```python
import time
```

```python
N = 1000000
list1 = list(range(N))
list2 = list(range(N))
```

```python
start = time.time()
list_add = [list1[i] + list2[i] for i in range(N)]
end = time.time()
print("List addition time:", end - start)
```

```python
np1 = np.arange(N)
np2 = np.arange(N)
```

```python
start = time.time()
np_add = np1 + np2
end = time.time()
print("NumPy addition time:", end - start)
```

2. Element-wise Multiplication

```python
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(N)]
end = time.time()
print("List multiplication time:", end - start)
```

```python
start = time.time()
np_mul = np1 * np2
end = time.time()
print("NumPy multiplication time:", end - start)
```

3. Dot Product

```python
start = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(N))
end = time.time()
print("List dot product time is given below:", end - start)
```

```python
start = time.time()
dot_np = np.dot(np1, np2)
end = time.time()
print("NumPy dot product time is given below:", end - start)
```

4. Matrix Multiplication
